```python
col_n = ['area','crime_rate','pm25','traffic','shockproof','school','age','floor']
```

```python
'''选取特征变量并划分数据集'''
X = df[col_n]  #选取特征变量
y =df.price  #设定price为y
```

```python
from sklearn.cross_validation import train_test_split #导入数据划分包 
# 把X、y转化为数组形式，以便于计算
X = np.array(X.values)  
y = np.array(y.values) 
#以25%的数据构建测试样本，剩余作为训练样本
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state =1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
'''建模之ridge回归'''
from sklearn import linear_model  #导入linear_model库
ridge= linear_model.Ridge(alpha=0.1)  # 设置lambda值
ridge.fit(X_train,y_train) #使用训练数据进行参数求解
```

```python
y_hat = ridge.predict(X_test) #对测试集的预测
from sklearn import metrics
# 用scikit-learn计算RMSE
print "RMSE_Ridge:", np.sqrt(metrics.mean_squared_error(y_test, y_hat))
```

```python
'''建模之lasso回归'''
from sklearn import linear_model  #导入linear_model库
lasso= linear_model.Lasso(alpha=0.1)  # 设置lambda值
lasso.fit(X_train,y_train) #使用训练数据进行参数求解
```

```python
y_hat_lasso = lasso.predict(X_test) #对测试集的预测
# 用scikit-learn计算RMSE
print "RMSE_losso:", np.sqrt(metrics.mean_squared_error(y_test, y_hat_lasso))
```

```python
'''建模之支持向量机回归'''
#从sklearn.svm中导入支持向量机回归模型SVR
from sklearn.svm import SVR
#使用线性核函数配置的支持向量机进行回归训练并预测
linear_svr = SVR(kernel='linear')
linear_svr.fit(X_train,y_train)  #训练模型
```

```python
y_hat_svr = linear_svr.predict(X_test) #对测试集的预测
# 用scikit-learn计算RMSE
print "RMSE_svr:", np.sqrt(metrics.mean_squared_error(y_test, y_hat_svr))
```

```python
'''建模之随机森林回归'''
#从sklearn中导入随机森林回归
from sklearn.ensemble import RandomForestRegressor
#1.使用随机森林进行回归训练并预测
rf = RandomForestRegressor(random_state=200,max_features=0.3)
rf.fit(X_train,y_train)
```

```python
y_hat_rf = rf.predict(X_test) #对测试集的预测
# 用scikit-learn计算RMSE
print "RMSE_rf:", np.sqrt(metrics.mean_squared_error(y_test, y_hat_rf))
```

```python
'''建模之SVR交叉验证'''
alphas_svr = np.linspace(0.1,1.2,20) #设置惩罚项C的参数序列
rmse_svr =[]  #设置rmse列表
for c in alphas_svr:
    model = SVR(kernel='linear',C=c)  #设定模型为svr
    model.fit(X_train,y_train) #使用训练数据进行参数求解
    y_hat = model.predict(X_test) #预测划分训练集
    rmse_svr.append(np.sqrt(metrics.mean_squared_error(y_test, y_hat)))
```

```python
plt.plot(alphas_svr,rmse_svr)  #绘制不同c取值的rmse结果
plt.title ('Cross Validation Score with Model SVR') #添加标题
plt.xlabel("alpha")  #添加x轴标签
plt.ylabel("rmse")  #添加y轴标签
plt.show()
```

```python
'''建模之lasso回归交叉验证'''
alphas_lasso = np.linspace(-0.1,0.1,20) #设置惩罚项alpha的参数序列
rmse_lasso =[] #设置rmse列表
for alpha in alphas_lasso:
    model = linear_model.Lasso(alpha) #设定模型为Lasso
    model.fit(X_train,y_train) #使用训练数据进行参数求解
    y_hat = model.predict(X_test) #预测划分训练集
    rmse_lasso.append(np.sqrt(metrics.mean_squared_error(y_test, y_hat))) 
```

```python
plt.plot(alphas_lasso,rmse_lasso) #绘制不同c取值的rmse结果
plt.title ('Cross Validation Score with Model Lasso') #添加标题
plt.xlabel("alpha") #添加x轴标签
plt.ylabel("rmse") #添加y轴标签
plt.show()
```
