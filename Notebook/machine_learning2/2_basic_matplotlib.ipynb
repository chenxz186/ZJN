```python
import matplotlib.pyplot as plt  #导入matplotlib库
import numpy as np  #导入numpy库
```

```python
x=np.arange(-np.pi,np.pi,0.1) #创建数组
y=np.sin(x)  #基于x创建正弦变化的y数组
```

```python
plt.plot(x,y)  #绘制图形
plt.show()  #显示图形
```

```python
plt.plot(x,y,color='green',linewidth=2.0, linestyle='-.')  #设置绿色、2像素、虚线线条
plt.show()  #显示图形
```

```python
sin,cos = np.sin(x), np.cos(x)
plt.plot(x,sin,label='sin')  #绘制图形
plt.plot(x,cos,color='red',linewidth=2.0,linestyle=':',label='cos') #绘制cosx图形
plt.grid(True)  #设置网格线
plt.legend() #设置图例
plt.show()
```

```python
plt.xlim(-4,4) # 设定横坐标范围
plt.ylim(-1.5,1.5)  # 设定纵坐标范围
plt.xlabel("x")  # 横轴标识
plt.ylabel("y") # 纵轴标识
plt.title("sinx plot") # 设定图形的标题
plt.plot(x,y)  #绘制图形
plt.show() 
```

```python
data = np.random.randint(1, 11, 4) #创建随机序列
labels = ['one','two','three','four'] #定义饼状图的标签，标签是列表
plt.pie(data,labels=labels)  # 绘制饼图
plt.axis('equal') # 设置x，y轴刻度一致，这样饼图才能是圆的
plt.legend()  #设置图例
plt.show()
```

```python
plt.scatter(x,y,color = 'r',marker = 'o')   # 绘制散点图
plt.show()
```

```python
bar1 = [1,2,3,4]  #创建数据
plt.bar(range(4), bar1) #绘制柱状图  
plt.show()  
```
