```python
import matplotlib.pyplot as plt  #导入matplotlib库
import numpy as np  #导入numpy库
import pandas as pd  #导入pandas库
from sklearn import model_selection  #模型比较和选择包
from sklearn.metrics import confusion_matrix #计算混淆矩阵，主要来评估分类的准确性
from sklearn.metrics import accuracy_score #计算精度得分
```

```python
df = pd.read_csv('..//z_data//7_traffic.csv') #读取csv数据
df.head()  #展示前五行数据
```

```python
len(df['traffic'])
```

```python
X = df.iloc[:,0:6]  #取df的前5列为X变量
y = df['traffic'] #设置y变量
```

```python
X.hist(xlabelsize=12,ylabelsize=12,figsize=(20,14))  #调整直方图尺寸
plt.show()
```

```python
y.hist(xlabelsize=12,ylabelsize=12,figsize=(8,5))  #调整直方图尺寸
plt.show()
```

```python
# 把X、y转化为数组形式，以便于计算
X = np.array(X.values)  
y = np.array(y.values)
```

```python
from sklearn.cross_validation import train_test_split  
#以25%的数据构建测试样本，剩余作为训练样本
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state =2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn import tree  #导入决策树库
tree_ID3 = tree.DecisionTreeClassifier(criterion='entropy') 
#默认采用的是gini，即是cart算法，在这里通过entropy设置ID3算法
tree_ID3 = tree_ID3.fit(X_train, y_train)  #采用ID3算法进行训练
y_pred_ID3 =tree_ID3.predict(X_test) #对测试集进行预测
accuracy_score(y_test, y_pred_ID3)  #计算准确率
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred_ID3)  #计算混淆矩阵
```

```python
from sklearn.ensemble import RandomForestClassifier #导入随机森林包
#定义一个随机森林分类器
clf = RandomForestClassifier(n_estimators=10,max_depth=None,
                             min_samples_split=2, oob_score=True, random_state=0)
clf.fit(X_train, y_train) #进行训练
```

```python
print clf.oob_score_ #袋外数据的预测精度
#可见此时我们的袋外分数有一定的提高。也就是时候模型的泛化能力增强了。
```

```python
y_pred_rf =clf.predict(X_test)  #预测测试集
y_pred_rf
```

```python
accuracy_score(y_test, y_pred_rf)  #计算准确率
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred_rf)  #计算混淆矩阵
```

```python
from sklearn.ensemble import ExtraTreesClassifier #导入极端森林包
#定义一个极端森林分类器
clf_extra = ExtraTreesClassifier(n_estimators=10, 
                                 max_depth=None,min_samples_split=2, random_state=0)
clf_extra.fit(X_train, y_train)  #进行训练
```

```python
y_pred_extra =clf_extra.predict(X_test)  #预测测试集
y_pred_extra
```

```python
accuracy_score(y_test, y_pred_extra)  #计算准确率
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred_extra)  #计算混淆矩阵
```
