```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
  

def show_pnl_single1():
    df1 = pd.read_excel('IO趋势15M每周核算数据.xls')
    df1.index = pd.to_datetime(df1['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式
    print(df1)

    fig = plt.figure(figsize=(8, 6), dpi=200)  # dpi是提高分辨率的参数
    ax1 = fig.add_subplot(111)
    df1['逐笔累计盈亏'].plot(ax=ax1, grid=True, color='blue', label='IO趋势S')
    plt.xlabel('日期')
    ax1.set_ylabel('IO趋势S')
    plt.legend(loc=2)
    plt.title('IO趋势S')
    plt.show()
  

def show_pnl_single2():
    df2 = pd.read_excel('IO趋势60M每周核算数据.xls')
    df2.index = pd.to_datetime(df2['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式
    print(df2)

    fig = plt.figure(figsize=(8, 6), dpi=200)  # dpi是提高分辨率的参数
    ax1 = fig.add_subplot(111)
    df2['逐笔累计盈亏'].plot(ax=ax1, grid=True, color='orange', label='IO趋势L')
    plt.xlabel('日期')
    ax1.set_ylabel('IO趋势L')
    plt.legend(loc=2)
    plt.title('IO趋势L')
    plt.show()
  

def show_pnl_single3(): 
    df3 = pd.read_excel('IO趋势1D每周核算数据.xls')
    df3.index = pd.to_datetime(df3['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式
    df3['3倍逐笔累计盈亏'] = df3['逐笔累计盈亏'] * 3
    print(df3)

    fig = plt.figure(figsize=(8, 6), dpi=200)  # dpi是提高分辨率的参数
    ax1 = fig.add_subplot(111)
    df3['3倍逐笔累计盈亏'].plot(ax=ax1, grid=True, color='purple', label='IO趋势1D')
    plt.xlabel('日期')
    ax1.set_ylabel('IO趋势1D')
    plt.legend(loc=2)
    plt.title('IO趋势1D')
    plt.show()
  

def show_pnl_7():
    df1 = pd.read_excel('IO趋势15M每周核算数据.xls')
    df1.index = pd.to_datetime(df1['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式

    df2 = pd.read_excel('IO趋势60M每周核算数据.xls')
    df2.index = pd.to_datetime(df2['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式
    
    df3 = pd.read_excel('IO趋势1D每周核算数据.xls')
    df3.index = pd.to_datetime(df3['交易日'], format='%Y-%m-%d')  # 字符串转换成日期格式
    df3['3倍逐笔累计盈亏'] = df3['逐笔累计盈亏'] * 3
    df3['IO三策略合并'] = df1['逐笔累计盈亏'] + df2['逐笔累计盈亏'] + df3['3倍逐笔累计盈亏']
    print(df3['IO三策略合并'])

    fig = plt.figure(figsize=(8, 6), dpi=200)  # dpi是提高分辨率的参数
    ax1 = fig.add_subplot(111)
    df1['逐笔累计盈亏'].plot(ax=ax1, grid=True, color='blue', label='IO趋势S')
    df2['逐笔累计盈亏'].plot(ax=ax1, grid=True, color='orange', label='IO趋势L')
    df3['3倍逐笔累计盈亏'].plot(ax=ax1, grid=True, color='purple', label='IO趋势1D')
    df3['IO三策略合并'].plot(ax=ax1, grid=True, color='red', label='IO三策略盈亏合并')
    plt.xlabel('日期')
    ax1.set_ylabel('IO三策略')
    plt.legend(loc=2)
    plt.title('IO三策略盈亏对比')
    plt.show()


if __name__ == '__main__':
    show_pnl_single1()
    show_pnl_single2()
    show_pnl_single3()
    show_pnl_7()
```
