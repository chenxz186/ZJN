```python
import easygui as g
import pyautogui as pg
import os
import time
import datetime


def statistics_time(func):
    """统计程序运行时间"""

    def test_time(*args, **kwargs):
        start_time = datetime.datetime.now()
        func(*args, **kwargs)
        end_time = datetime.datetime.now()
        result = end_time - start_time
        print('\n\n程序开始时间{}，\n结束时间{}，\n耗时{}'.format(start_time, end_time, result))

    return test_time


def run(x, y, contract, start_price, end_price, root_price):
    dir_path = 'F:\\杂物放置处\\商品期权'
    # 对T型报价图截取
    im = pg.screenshot()
    string = '{}-{}-7型报价图.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    print(string)
    time.sleep(1)
    # 对期货价进行截图
    pg.doubleClick(850, 60)
    time.sleep(1)
    im = pg.screenshot()
    string = '{}-{}.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    pg.keyDown('esc')
    pg.keyUp('esc')
    print(string)
    # 上面是对期货价进行截图

    commodity_type = contract.split(contract[-4:])[0]  # 获取商品代码，例如m2101提取后是m
    space_num = 0
    # 铁矿石
    if commodity_type == 'i':
        if start_price <= 300:
            space_num += (start_price - root_price) / 5
        elif 300 < start_price <= 1000:
            if root_price <= 300:
                space_num += (300 - root_price) / 5 + (start_price - 300) / 10
            elif 300 < root_price <= start_price:
                space_num += (start_price - root_price) / 10
        elif start_price > 1000:
            if root_price <= 300:
                space_num += (300 - root_price) / 5 + (1000 - 300) / 10 + (start_price - 1000) / 20
            elif 300 < root_price <= 1000:
                space_num += (1000 - root_price) / 10 + (start_price - 1000) / 20
            elif 1000 < root_price <= start_price:
                space_num += (start_price - root_price) / 20
    # 豆粕
    if commodity_type == 'm':
        if start_price <= 2000:
            space_num += (start_price - root_price) / 25
        elif 2000 < start_price <= 5000:
            if root_price <= 2000:
                space_num += (2000 - root_price) / 25 + (start_price - 2000) / 50
            elif 2000 < root_price <= start_price:
                space_num += (start_price - root_price) / 50
        elif start_price > 5000:
            if root_price <= 2000:
                space_num += (2000 - root_price) / 25 + (5000 - 2000) / 50 + (start_price - 5000) / 100
            elif 2000 < root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (start_price - 5000) / 100
            elif 5000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
    # 液化气
    if commodity_type == 'pg':
        if start_price <= 2000:
            space_num += (start_price - root_price) / 25
        elif 2000 < start_price <= 6000:
            if root_price <= 2000:
                space_num += (2000 - root_price) / 25 + (start_price - 2000) / 50
            elif 2000 < root_price <= start_price:
                space_num += (start_price - root_price) / 50
        elif start_price > 6000:
            if root_price <= 2000:
                space_num += (2000 - root_price) / 25 + (6000 - 2000) / 50 + (start_price - 6000) / 100
            elif 2000 < root_price <= 6000:
                space_num += (6000 - root_price) / 50 + (start_price - 6000) / 100
            elif 6000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
    # pp、pvc、塑料
    if commodity_type == 'pp' or commodity_type == 'v' or commodity_type == 'l':
        if start_price <= 5000:
            space_num += (start_price - root_price) / 50
        elif 5000 < start_price <= 10000:
            if root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (start_price - 5000) / 100
            elif 5000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
        elif start_price > 10000:
            if root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (10000 - 5000) / 100 + (start_price - 10000) / 200
            elif 5000 < root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 200
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
    y = y + space_num * 22
    while True:
        # 截取认购
        pg.doubleClick(x, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-C-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        # 截取认沽
        pg.doubleClick(x + 1200, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-P-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        if start_price == end_price:
            break
        # 铁矿石
        if commodity_type == 'i':
            if start_price <= 300:
                start_price += 5
            elif 300 < start_price <= 1000:
                start_price += 10
            elif start_price > 1000:
                start_price += 20
        # 豆粕
        if commodity_type == 'm':
            if start_price <= 2000:
                start_price += 25
            elif 2000 < start_price <= 5000:
                start_price += 50
            elif start_price > 5000:
                start_price += 100
        # 液化气
        if commodity_type == 'pg':
            if start_price <= 2000:
                start_price += 25
            elif 2000 < start_price <= 6000:
                start_price += 50
            elif start_price > 6000:
                start_price += 100
        # pp, pvc, 塑料
        if commodity_type == 'pp' or commodity_type == 'v' or commodity_type == 'l':
            if start_price <= 5000:
                start_price += 50
            elif 5000 < start_price <= 10000:
                start_price += 100
            elif start_price > 10000:
                start_price += 200
        y += 22


@statistics_time
def main():
    if 'i' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 100)
        time.sleep(1)
        pg.click(430, 80)
        time.sleep(1)
        pg.click(400, 240)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(1)
        run(400, 100, 'i2102', 950, 1000, 650)
    if 'l' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 100)
        time.sleep(1)
        pg.click(430, 80)
        time.sleep(1)
        pg.click(400, 260)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(1)
        run(400, 100, 'l2102', 7700, 8000, 5800)
    if 'm' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 200)
        time.sleep(1)
        pg.click(430, 260)
        time.sleep(1)
        pg.click(400, 80)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(1)
        run(400, 100, 'm2103', 3000, 3200, 2400)
    if 'pg' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 200)
        time.sleep(1)
        pg.click(430, 260)
        time.sleep(1)
        pg.click(400, 100)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'pg2102', 3800, 3900, 2750)
    if 'pp' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 200)
        time.sleep(1)
        pg.click(430, 260)
        time.sleep(1)
        pg.click(400, 120)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'pp2102', 8200, 8500, 6100)
    if 'v' in target:
        pg.click(400, 60)
        time.sleep(1)
        pg.click(430, 200)
        time.sleep(1)
        pg.click(430, 260)
        time.sleep(1)
        pg.click(400, 140)
        time.sleep(1)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'v2102', 7400, 7600, 5300)


date = g.enterbox('请输入日期，例如2020年7月20日，即输入20200720', '日期输入')
if date is None:
    exit()

target = [
    'i',
    'm',
    'pg',
    'pp',
    'v',
    'l'
]
if __name__ == '__main__':
    main()
```
