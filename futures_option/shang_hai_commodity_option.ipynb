```python
import easygui as g
import pyautogui as pg
import os
import time
import datetime


def statistics_time(func):
    """统计程序运行时间"""

    def test_time(*args, **kwargs):
        start_time = datetime.datetime.now()
        func(*args, **kwargs)
        end_time = datetime.datetime.now()
        result = end_time - start_time
        print('\n\n程序开始时间{}，\n结束时间{}，\n耗时{}'.format(start_time, end_time, result))

    return test_time


def run(x, y, contract, start_price, end_price, root_price):
    dir_path = 'F:\\杂物放置处\\商品期权'
    # 对T型报价图截取
    im = pg.screenshot()
    string = '{}-{}-7型报价图.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    print(string)
    time.sleep(1)
    # 对期货价进行截图
    pg.doubleClick(850, 60)
    time.sleep(1)
    im = pg.screenshot()
    string = '{}-{}.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    pg.keyDown('esc')
    pg.keyUp('esc')
    print(string)
    # 上面是对期货价进行截图

    commodity_type = contract.split(contract[-4:])[0]  # 获取商品代码，例如m2101提取后是m
    space_num = 0
    # 铝期权
    if commodity_type == 'al':
        if start_price <= 10000:
            space_num += (start_price - root_price) / 50
        elif 10000 < start_price <= 20000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 50 + (start_price - 10000) / 100
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
        elif start_price > 20000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 50 + (20000 - 10000) / 100 + (start_price - 20000) / 200
            elif 10000 < root_price <= 20000:
                space_num += (20000 - root_price) / 100 + (start_price - 20000) / 200
            elif 20000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
    # 铜期权
    if commodity_type == 'cu':
        if start_price <= 40000:
            space_num += (start_price - root_price) / 500
        elif 40000 < start_price <= 80000:
            if root_price <= 40000:
                space_num += (40000 - root_price) / 500 + (start_price - 40000) / 1000
            elif 40000 < root_price <= start_price:
                space_num += (start_price - root_price) / 1000
        elif start_price > 80000:
            if root_price <= 40000:
                space_num += (40000 - root_price) / 500 + (80000 - 40000) / 1000 + (start_price - 80000) / 2000
            elif 40000 < root_price <= 80000:
                space_num += (80000 - root_price) / 1000 + (start_price - 80000) / 2000
            elif 80000 < root_price <= start_price:
                space_num += (start_price - root_price) / 2000
    # 锌期权
    if commodity_type == 'zn':
        if start_price <= 10000:
            space_num += (start_price - root_price) / 100
        elif 10000 < start_price <= 25000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 200
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
        elif start_price > 25000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (25000 - 10000) / 200 + (start_price - 25000) / 500
            elif 10000 < root_price <= 25000:
                space_num += (25000 - root_price) / 200 + (start_price - 25000) / 500
            elif 25000 < root_price <= start_price:
                space_num += (start_price - root_price) / 500
    # 橡胶期权
    if commodity_type == 'ru':
        if start_price <= 10000:
            space_num += (start_price - root_price) / 100
        elif 10000 < start_price <= 25000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 250
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 250
        elif start_price > 25000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (25000 - 10000) / 250 + (start_price - 25000) / 500
            elif 10000 < root_price <= 25000:
                space_num += (25000 - root_price) / 250 + (start_price - 25000) / 500
            elif 25000 < root_price <= start_price:
                space_num += (start_price - root_price) / 500
    y = y + space_num * 22
    while True:
        # 截取认购
        pg.doubleClick(x, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-C-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        # 截取认沽
        pg.doubleClick(x + 1200, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-P-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        if start_price == end_price:
            break
        # 铝期权
        if commodity_type == 'al':
            if start_price <= 10000:
                start_price += 50
            elif 10000 < start_price <= 20000:
                start_price += 100
            elif start_price > 20000:
                start_price += 200
        # 铜期权
        if commodity_type == 'cu':
            if start_price <= 40000:
                start_price += 500
            elif 40000 < start_price <= 80000:
                start_price += 1000
            elif start_price > 80000:
                start_price += 2000
        # 锌期权
        if commodity_type == 'zn':
            if start_price <= 10000:
                start_price += 100
            elif 10000 < start_price <= 25000:
                start_price += 200
            elif start_price > 25000:
                start_price += 500
        # 橡胶期权
        if commodity_type == 'ru':
            if start_price <= 10000:
                start_price += 100
            elif 10000 < start_price <= 25000:
                start_price += 250
            elif start_price > 25000:
                start_price += 500
        y += 22


@statistics_time
def main():
    if 'al' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 100)
        time.sleep(0.5)
        pg.click(430, 80)
        time.sleep(0.5)
        pg.click(400, 140)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(0.5)
        run(400, 100, 'al2101', 16200, 16800, 14400)
    if 'cu' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 100)
        time.sleep(0.5)
        pg.click(430, 80)
        time.sleep(0.5)
        pg.click(400, 160)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(0.5)
        run(400, 100, 'cu2101', 58000, 60000, 34000)
    if 'zn' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 100)
        time.sleep(0.5)
        pg.click(430, 80)
        time.sleep(0.5)
        pg.click(400, 170)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(0.5)
        run(400, 100, 'zn2101', 21600, 22400, 16200)
    if 'ru' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 100)
        time.sleep(0.5)
        pg.click(430, 80)
        time.sleep(0.5)
        pg.click(400, 200)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'ru2101', 14250, 15000, 9400)


date = g.enterbox('请输入日期，例如2020年7月20日，即输入20200720', '日期输入')
if date is None:
    exit()

target = [
    'al',
    'cu',
    'zn',
    'ru'
]
if __name__ == '__main__':
    main()
```
