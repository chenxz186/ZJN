```python
import easygui as g
import pyautogui as pg
import os
import time
import datetime


def statistics_time(func):
    """统计程序运行时间"""

    def test_time(*args, **kwargs):
        start_time = datetime.datetime.now()
        func(*args, **kwargs)
        end_time = datetime.datetime.now()
        result = end_time - start_time
        print('\n\n程序开始时间{}，\n结束时间{}，\n耗时{}'.format(start_time, end_time, result))

    return test_time


def run(x, y, contract, start_price, end_price, root_price):
    dir_path = 'F:\\杂物放置处\\商品期权'
    # 对T型报价图截取
    im = pg.screenshot()
    string = '{}-{}-7型报价图.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    print(string)
    time.sleep(1)
    # 对期货价进行截图
    pg.doubleClick(850, 60)
    time.sleep(1)
    im = pg.screenshot()
    string = '{}-{}.jpg'.format(date, contract)
    filename = dir_path + '/' + string
    im.save(filename)
    pg.keyDown('esc')
    pg.keyUp('esc')
    print(string)
    # 上面是对期货价进行截图

    commodity_type = contract.split(contract[-4:])[0]  # 获取商品代码，例如m2101提取后是m
    space_num = 0
    # 甲醇、菜粕
    if commodity_type == 'ma' or commodity_type == 'rm':
        if start_price <= 2500:
            space_num += (start_price - root_price) / 25
        elif 2500 < start_price <= 5000:
            if root_price <= 2500:
                space_num += (2500 - root_price) / 25 + (start_price - 2500) / 50
            elif 2500 < root_price <= start_price:
                space_num += (start_price - root_price) / 50
        elif start_price > 5000:
            if root_price <= 2500:
                space_num += (2500 - root_price) / 25 + (5000 - 2500) / 50 + (start_price - 5000) / 100
            elif 2500 < root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (start_price - 5000) / 100
            elif 5000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
    # 动力煤
    if commodity_type == 'zc':
        if start_price <= 500:
            space_num += (start_price - root_price) / 5
        elif start_price > 500:
            if root_price <= 500:
                space_num += (500 - root_price) / 5 + (start_price - 500) / 10
            elif 500 < root_price <= start_price:
                space_num += (start_price - root_price) / 10
    # PTA
    if commodity_type == 'pta':
        if start_price <= 5000:
            space_num += (start_price - root_price) / 50
        elif 5000 < start_price <= 10000:
            if root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (start_price - 5000) / 100
            elif 5000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
        elif start_price > 10000:
            if root_price <= 5000:
                space_num += (5000 - root_price) / 50 + (10000 - 5000) / 100 + (start_price - 10000) / 200
            elif 5000 < root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 200
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
    # 白糖
    if commodity_type == 'sr':
        if start_price <= 3000:
            space_num += (start_price - root_price) / 50
        elif 3000 < start_price <= 10000:
            if root_price <= 3000:
                space_num += (3000 - root_price) / 50 + (start_price - 3000) / 100
            elif 3000 < root_price <= start_price:
                space_num += (start_price - root_price) / 100
        elif start_price > 10000:
            if root_price <= 3000:
                space_num += (3000 - root_price) / 50 + (10000 - 3000) / 100 + (start_price - 10000) / 200
            elif 3000 < root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 200
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
    # 棉花
    if commodity_type == 'cf':
        if start_price <= 10000:
            space_num += (start_price - root_price) / 100
        elif 10000 < start_price <= 20000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (start_price - 10000) / 200
            elif 10000 < root_price <= start_price:
                space_num += (start_price - root_price) / 200
        elif start_price > 20000:
            if root_price <= 10000:
                space_num += (10000 - root_price) / 100 + (20000 - 10000) / 200 + (start_price - 20000) / 400
            elif 10000 < root_price <= 20000:
                space_num += (20000 - root_price) / 200 + (start_price - 20000) / 400
            elif 20000 < root_price <= start_price:
                space_num += (start_price - root_price) / 400
    y = y + space_num * 22
    while True:
        # 截取认购
        pg.doubleClick(x, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-C-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        # 截取认沽
        pg.doubleClick(x + 1200, y)
        time.sleep(1)
        im = pg.screenshot()
        string = '{}-{}-P-{}.jpg'.format(date, contract, start_price)
        filename = dir_path + '/' + string
        im.save(filename)
        pg.keyDown('esc')
        pg.keyUp('esc')
        print(string)

        if start_price == end_price:
            break
        # 甲醇、菜粕
        if commodity_type == 'ma' or commodity_type == 'rm':
            if start_price <= 2500:
                start_price += 25
            elif 2500 < start_price <= 5000:
                start_price += 50
            elif start_price > 5000:
                start_price += 100
        # 动力煤
        if commodity_type == 'zc':
            if start_price <= 500:
                start_price += 5
            elif start_price > 500:
                start_price += 10
        # PTA
        if commodity_type == 'pta':
            if start_price <= 5000:
                start_price += 50
            elif 5000 < start_price <= 10000:
                start_price += 100
            elif start_price > 10000:
                start_price += 200
        # 白糖
        if commodity_type == 'sr':
            if start_price <= 3000:
                start_price += 50
            elif 3000 < start_price <= 10000:
                start_price += 100
            elif start_price > 10000:
                start_price += 200
        # 棉花
        if commodity_type == 'cf':
            if start_price <= 10000:
                start_price += 100
            elif 10000 < start_price <= 20000:
                start_price += 200
            elif start_price > 20000:
                start_price += 400
        y += 22


@statistics_time
def main():
    if 'ma' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 170)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(0.5)
        run(400, 100, 'ma2102', 2250, 2350, 1825)
    if 'zc' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 240)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(1)
        run(400, 100, 'zc2102', 710, 730, 510)
    if 'pta' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 220)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        time.sleep(0.5)
        run(400, 100, 'pta2103', 3500, 3700, 3000)
    if 'rm' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 180)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'rm2102', 2400, 2500, 2025)
    if 'cf' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 150)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'cf2102', 14000, 14800, 11000)
    if 'sr' in target:
        pg.click(400, 60)
        time.sleep(0.5)
        pg.click(430, 200)
        time.sleep(0.5)
        pg.click(430, 260)
        time.sleep(0.5)
        pg.click(400, 200)
        time.sleep(0.5)
        pg.click(2525, 980)  # 有时T型报价图过长，出现滚动条，根据实际滚动向上向下，这里向下滚动。
        run(400, 100, 'sr2102', 4800, 5200, 4300)


date = g.enterbox('请输入日期，例如2020年7月20日，即输入20200720', '日期输入')
if date is None:
    exit()

target = [
    'ma',
    'zc',
    'pta',
    'rm',
    'cf',
    'sr'
]
if __name__ == '__main__':
    main()
```
