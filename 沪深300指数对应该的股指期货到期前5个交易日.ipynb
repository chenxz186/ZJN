```python
import tushare as ts

pro = ts.pro_api('126866eac3360351608f628643fa64c69002f8c593ba6822fb7957a0')
df = pro.index_daily(ts_code='000300.SH', start_date='20210816', end_date='20210820')
df.pop('vol')
df.pop('amount')
new_names = {'ts_code': '股票代码', 'trade_date': '交易日', 'close': '收盘价', 'open': '开盘价', 'high': '最高价', 'low': '最低价', 'pre_close': '昨收盘价', 'change': '涨跌', 'pct_chg': '涨幅(%)'}
df = df.rename(columns=new_names)
column_names = list(df)
for each in column_names[2:]:
    df[each] = round(df[each], 2)
new_df = df.drop([1, 2, 3, 4])
new_df['开盘价'] = df.iloc[4]['开盘价']
new_df['最高价'] = max(df['最高价'])
new_df['最低价'] = min(df['最低价'])
new_df['昨收盘价'] = df.iloc[4]['昨收盘价']
new_df['涨跌'] = df.iloc[0]['收盘价'] - df.iloc[4]['昨收盘价']
new_df['涨幅(%)'] = round((new_df['涨跌'] / new_df['昨收盘价']) * 100, 2)
df.pop('昨收盘价')
print(df)
new_df
```
